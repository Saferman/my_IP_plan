数据透视表（Pivot Table）是一种常用的数据汇总与分析工具，最典型的应用场景是在 Excel 和 Pandas（Python） 里。它的作用就是：

快速汇总：把原始明细表（往往行数很多）中的数据，按某些字段进行分类、分组和统计。

灵活分析：可以随意拖动“行、列、值、筛选”，生成不同的统计报表。

多维度交叉分析：比如既可以按“地区”统计销售额，也可以再按“年份”细分，形成交叉表

### 1 - 加载数据

读取当前目录下 `"某超市销售数据.csv"` 并设置千分位符号为 `,`

```python
import pandas as pd
df = pd.read_csv("某超市销售数据.csv", thousands=',')
df.pivot_table(values='销售额', index='省/自治区')
```

### 2 - 数据透视｜默认

制作各省「平均销售额」的数据透视表

```python
pivot_2 = df.pivot_table(values='销售额', index='省/自治区')
print(pivot_2.round(2))
print("\n" + "-"*50 + "\n")
```

### 3 - 数据透视｜指定方法

制作各省「销售总额」的数据透视表

```python
df.pivot_table(values='销售额', index='省/自治区', aggfunc='sum')
```
